```python
plt.figure(figsize = (10,10))
sns.heatmap(df.corr(), cmap = sns.cm.rocket_r)
plt.xticks(rotation = 60)
plt.savefig("corr_heatmap")
```

```python
## check those interesting flats!  
# No bedrooms but has bathrooms
df.loc[(df.bedrooms != 0) & (df.bathrooms == 0) ,]
```

```python
# No bathrooms but has bedrooms
df.loc[(df.bedrooms == 0) & (df.bathrooms != 0) ,]
```

```python
# Neither bathrooms nor bedrooms
df.loc[(df.bedrooms == 0) & (df.bathrooms == 0) ,]
```

```python
# Two flats with absurd bedroom numbers, but low prices and small sqft living!

df.loc[(df.bedrooms == 11) | (df.bedrooms == 33), ]
```

```python
df.loc[~((df.bedrooms == 11) | (df.bedrooms == 33) | (df.bedrooms == 0) | (df.bathrooms == 0)), :]
```

```python
df.zipcode.value_counts()
```

```python
df.drop(["lat", "long"], axis = 1, inplace = True)
```

```python
df["renovated"] = df.yr_renovated.apply(lambda x: 0 if x == 0 else 1)
```

```python
df["avg_price_per_sqft"] = df.price/df.sqft_living
```

```python
df["sqft_living_percent"] = df.sqft_living/df.sqft_lot
```
